## Ejercicio 3

Define una función llamada is_romano() dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano.

```python
class Numero:
    def __init__(self, normal):
        self.normal = normal
        self.romano = self.de_entero_a_romano(normal)
    def de_entero_a_romano(self,numero):
        romanos = {1000 :'M', 900 :'CM', 500 : 'D', 400 :'CD', 100 : 'C', 90 : 'XC', 50 : 'L', 40 : 'XL', 10 :'X', 9 : 'IX', 5 : 'V', 4 : 'IV', 1 :'I'}
        resultado = ''
        for valor, letra in romanos.items():
            while numero >= valor:
                resultado += letra
                numero -= valor
        return resultado

    def de_romano_a_entero(self,numero_romano):
        dicc_romano = {'I': 1, 'V': 5, 'IX' : 9 , 'X': 10, 'L' : 50, 'C': 100, 'D' : 500, 'M' : 1000}
        total = 0
        previo = 0
        for i in range(len(numero_romano)-1,-1,-1):
            actual = dicc_romano[numero_romano[i]]
            if actual >= previo:
                total += actual
            else:
                total -= actual
            previo = actual
        return total 
    def imprime(self):
        numero_normal = str(self.normal)
        numero_romano = self.romano
        final = numero_normal + ' ' + numero_romano
        print(final)
    def suma_romano(self, numero_nuevo):
        if(isinstance(numero_nuevo, str)):
            numero = self.de_romano_a_entero(numero_nuevo)
            suma = self.normal + numero
            self.normal = suma
            self.romano = self.de_entero_a_romano(suma)
        return suma
    def is_romano(self, numero):
        valores_romanos = {'I': 1, 'V': 5, 'X': 10, 'L' : 50, 'C': 100, 'D' : 500, 'M' : 1000}
        inicial = 'I'
        for actual in numero:
            if actual not in valores_romanos:
                return False
            if valores_romanos[actual] > valores_romanos[inicial] and inicial != 'I' and inicial != 'X' and inicial != 'C':
                return False
            if numero.count(actual) > 3 and actual not in ['V', 'L', 'D']:
                return False
            inicial = actual
        return True
    
```
